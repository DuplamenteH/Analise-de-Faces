```
#carregando bibliotecas e as imagens
import cv2 
import matplotlib.pyplot as plt
```

```
imagem = cv2.imread('/content/drive/My Drive/DadosAlura/analises de faces/imagens/px-people.jpg')
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
```

```
plt.figure(figsize=(20,10))
plt.imshow(imagem)
```

```
#Deixando a imagem cinza
imagem_gray = cv2.cvtColor(imagem, cv2.COLOR_RGB2GRAY)
```

```
#classificador
classicador_img = cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
```

```
faces = classicador_img.detectMultiScale(imagem_gray,1.3,6)
```

```
len(faces)
```

```
faces[0]
```

```
#plotando a imagem com as marcações

imagem_copia =  imagem.copy()
```

```
for (x,y,w,h) in faces :
    cv2.rectangle(imagem_copia,(x,y), (x+w, y+h),(0,0,255),3)
```

```
plt.figure(figsize=(20,10))
plt.imshow(imagem_copia)
```

```
#cortando as imagens
face_img = 0
for (x,y,w,h) in faces :
    face_img+=1
    img_recortada = imagem[y:y+h,x:x+w]
    img_recortada = cv2.cvtColor(img_recortada, cv2.COLOR_RGB2BGR)
    cv2.imwrite("face_"+ str(face_img) + ".png",img_recortada)
```
